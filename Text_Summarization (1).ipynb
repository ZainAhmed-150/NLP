```python
#importing all the necessary libraries
import nltk
from nltk.corpus import stopwords
from string import punctuation
from nltk.collocations import *
from nltk.tokenize import sent_tokenize as st, word_tokenize as wt
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
#Method to automate above two cells
def get_text(article_link):
    page = urlopen(article_link).read().decode('utf8','ignore')
    soup = BeautifulSoup(page,'lxml')
    text = ' '.join(map(lambda p: p.text, soup.find_all('p')))
    return text
```

```python
#Getting text
text = get_text(article_link)
text
```

```python
#Tokenizing sentences
sents = st(text.lower())
sents
```

```python
#Tokenizing words
words = wt(text.lower())
words
```

```python
#Creating set of stopwords
stopwords_set = set(nltk.corpus.stopwords.words('english') + list(punctuation))
stopwords_set
```

```python
#Removing stopwords
words = [word for word in words if word not in stopwords_set]
words
```

```python
#Importing FreqDist
from nltk.probability import FreqDist
```

```python
#Calculating word frequency
freq = FreqDist(words)
freq
```

```python
#Getting n most - frequent words
from heapq import nlargest
nlargest(10, freq, key = freq.get)
```

```python
#Creating sentence ranking dictionary based on word frequency
from collections import defaultdict
ranking = defaultdict(int)

for i,sent in enumerate(sents):
    for w in wt(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]

ranking
```

```python
#Getting n highest ranked sentences
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

```python
#Getting n highest ranked sentences in ascending order - to get the right context
[sents[idx] for idx in sorted(sents_idx)]
```

```python
#Automating all above steps
def summarize_text(article_link,n):
    
    #downloading the article
    page = urlopen(article_link).read().decode('utf8','ignore')
    soup = BeautifulSoup(page,'lxml')
    
    #accessing all paragraphs
    text = ' '.join(map(lambda p: p.text, soup.find_all('p')))
    
    #tokenization of text corpus into list of sentences
    sents = st(text.lower())
    
    assert n <= len(sents)
    
    #tokenization of sentences into words
    words = wt(text.lower())
    _stopwords = set(nltk.corpus.stopwords.words('english') + list(punctuation))
    
    #words left after removing stopwords
    words = [word for word in words if word not in _stopwords]
    
    #calculating word frequency
    freq = FreqDist(words)
    
    #creating a default dict to store the ranking of sentences in descending order of sum of frequency of all words in it
    ranking = defaultdict(int)
    
    for i,sent in enumerate(sents):
        for w in wt(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
    
    #creating summary with top n most weighted(significant) sentences in correct order
    sents_idx = nlargest(n,ranking,key=ranking.get)
    
    return [sents[idx] for idx in sorted(sents_idx)]
```

```python
#Fetching article summary
article_link = 'https://en.wikipedia.org/wiki/Main_Page'
summarize_text(article_link,4)
```
